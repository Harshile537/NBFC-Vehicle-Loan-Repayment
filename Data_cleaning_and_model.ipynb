Converting above col. numeric


```python
#Replacing $ in data to nan and then converting to numeric
df_train.Client_Income.replace("$", np.nan, inplace = True)
df_train.Client_Income = pd.to_numeric(df_train.Client_Income)
```

```python
df_train['Client_Income'].info()
```

```python
df_train.Credit_Amount.replace("$", np.nan, inplace = True)
df_train.Credit_Amount = pd.to_numeric(df_train.Credit_Amount)
```

```python
df_train.Loan_Annuity.replace(["#VALUE!","$"], np.nan, inplace = True)
df_train.Loan_Annuity = pd.to_numeric(df_train.Loan_Annuity)
```

```python
df_train.Age_Days.replace("x", np.nan, inplace = True)
df_train.Age_Days = pd.to_numeric(df_train.Age_Days)
```

```python
df_train.Employed_Days.replace("x", np.nan, inplace = True)
df_train.Employed_Days = pd.to_numeric(df_train.Employed_Days)
```

```python
df_train.Registration_Days.replace("x", np.nan, inplace = True)
df_train.Registration_Days = pd.to_numeric(df_train.Registration_Days)
```

```python
df_train.ID_Days.replace("x", np.nan, inplace = True)
df_train.ID_Days = pd.to_numeric(df_train.ID_Days)
```

```python
df_train["Score_Source_3"].replace("&", np.nan, inplace=True)
df_train.Score_Source_3 = pd.to_numeric(df_train.Score_Source_3)
```

```python

# New Column is created on the basis of Population_Region_Relative column whether the Client area is Populated area or not

df_train["Population_Region_Relative"].replace(["#","@"], np.nan, inplace=True)
df_train.Population_Region_Relative = pd.to_numeric(df_train.Population_Region_Relative)
populated_Region_mean = df_train.Population_Region_Relative.mean()
temp = df_train.copy()
temp[temp["Population_Region_Relative"] < populated_Region_mean] = 0 # 0 for less populated area
temp[temp["Population_Region_Relative"] >= populated_Region_mean] = 1 # 1 for high populated area
df_train["Population_Region_Relative"] = temp["Population_Region_Relative"]
```

```python
df_train.head(3)
```

```python
#dropping columns with more than 50% null value
df_train.drop(columns=["Own_House_Age","Score_Source_1","Social_Circle_Default"],axis=1,inplace=True)
df_train.head(3)
```

```python
df_train.info()
```

```python
from sklearn.impute import SimpleImputer

si_mean = SimpleImputer(strategy="mean", missing_values=np.nan)
si_mode = SimpleImputer(strategy="most_frequent", missing_values=np.nan)
```

```python
# Applying Mode
df_train.iloc[:, 0 : 5] = si_mode.fit_transform(df_train.iloc[:, 0 : 5])
df_train.iloc[:, 8 : 16] = si_mode.fit_transform(df_train.iloc[:, 8 : 16])
df_train.iloc[:, 20 : 31] = si_mode.fit_transform(df_train.iloc[:, 20 : 31])
df_train.iloc[:, 34 : 35] = si_mode.fit_transform(df_train.iloc[:, 34 : 35])
```

```python
# Applying Mean
df_train.iloc[:, 6 : 8] = si_mean.fit_transform(df_train.iloc[:, 6 : 8])
df_train.iloc[:, 31 : 34] = si_mean.fit_transform(df_train.iloc[:, 31 : 34])
```

```python
# Applying Mean with Round Function for req columns
df_train.iloc[:, 5 : 6] = np.round(si_mean.fit_transform(df_train.iloc[:, 5 : 6]))
df_train.iloc[:, 16 : 20] = np.round(si_mean.fit_transform(df_train.iloc[:, 16 : 20]))
```

```python
# Label encoding for categorical columns
from sklearn.preprocessing import LabelEncoder
cols = ["Accompany_Client", "Client_Marital_Status", "Client_Income_Type", "Client_Education", "Client_Gender", "Loan_Contract_Type", "Client_Housing_Type", "Client_Occupation", "Client_Permanent_Match_Tag", "Client_Contact_Work_Tag", "Type_Organization"]
df_train[cols] = df_train[cols].apply(LabelEncoder().fit_transform)
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score,auc,roc_curve,f1_score, recall_score, precision_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
import pickle
from imblearn.combine import SMOTETomek
```

```python
X = df_train.drop("Default", axis = 1)
y = df_train["Default"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
rf_classifier = RandomForestClassifier()
rf_classifier.fit(X_train, y_train)
y_pred = rf_classifier.predict(X_test)
accuracy_score(y_test, y_pred) * 100
```

```python
#training_score = []
#testing_score = []

def accuracy_check(x,y,models):
    #train test split

    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)

    model=models().fit(x_train,y_train)

    y_trainpred=model.predict(x_train)
    y_testpred=model.predict(x_test)

    #checking the accuracy_score
    accuracy_train=accuracy_score(y_train,y_trainpred)
    accuracy_test=accuracy_score(y_test,y_testpred)
    Precision_score=precision_score(y_test,y_testpred)
    Recall_score=recall_score(y_test,y_testpred)
    F1_Score=f1_score(y_test,y_testpred)

    metrics={"Model": models.__name__,
            "Accuracy_Train":accuracy_train,
            "Accuracy_Test":accuracy_test,
            "Precision_score":Precision_score,
             "Recall_score":Recall_score,
             "F1_score":F1_Score
            }
    return metrics
```

```python
accuracy_check(X,y,LogisticRegression)
```

```python
accuracy_check(X,y,KNeighborsClassifier)
```

```python
accuracy_check(X,y,DecisionTreeClassifier)
```

```python
accuracy_check(X,y,RandomForestClassifier)
```

```python
accuracy_check(X,y,AdaBoostClassifier)
```

```python
X_train,x_test,y_train,y_test=train_test_split(x_new,y_new,test_size=0.2,random_state=42)

model=RandomForestClassifier().fit(X_train,y_train)

y_pred_train=model.predict(X_train)
y_pred_test=model.predict(x_test)

#checking the accuracy score of train and test data

accuracy_train=accuracy_score(y_train,y_pred_train)
accuracy_test=accuracy_score(y_test,y_pred_test)

print("Accuracy scores of train and test data")
print("Train_Accuracy",accuracy_train)
print("Test_Accuracy",accuracy_test)
print("Confusion_matrix for Test")
print(confusion_matrix(y_true= y_test, y_pred= y_pred_test))
```
